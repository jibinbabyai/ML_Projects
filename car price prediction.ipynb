```python
dfe= pd.get_dummies(df[["Location","Fuel_Type","Transmission","Owner_Type","Brand"]],drop_first= True,dtype=int)
dfe
```

```python
df_1 = pd.concat([df,dfe],axis=1)
df_1.head()
```

```python
df_1=df_1.drop(["Unnamed: 0","Name","Location","Fuel_Type","Transmission","Owner_Type","Brand","New_Price"],axis=1)
df_1.head()
```

```python
df_1.shape
```

```python
df_1["Mileage"] = df_1["Mileage"].str.replace("kmpl"," ")
df_1["Mileage"] = df_1["Mileage"].str.replace("km/kg"," ")
df_1["Engine"] = df_1["Engine"].str.replace("CC"," ")
df_1["Power"] = df_1["Power"].str.replace("bhp"," ")
```

```python
df_1["Mileage"] = df_1["Mileage"].str.replace("null","0")
df_1["Mileage"] = df_1["Mileage"].str.replace("null","0")
df_1["Engine"] = df_1["Engine"].str.replace("null","0")
df_1["Power"] = df_1["Power"].str.replace("null","0")
```

```python
# converting data type
df_1["Engine"] = df_1["Engine"].astype(float)
df_1["Mileage"] = df_1["Mileage"].astype(float)
df_1["Power"] = df_1["Power"].astype(float)
df_1.dtypes
```

```python
# treating 0 to missing value (engine, mileage can't be as 0)

# mileage,engine,power ==>null ====>"0"
# mileage,engine,power  ===>numerical  ===>0 ===>NaN

df_1.loc[df_1.Engine==0,"Engine"]=np.nan
df_1.loc[df_1.Power==0,"Power"]=np.nan
df_1.loc[df_1.Mileage==0,"Mileage"]=np.nan

```

```python
df_1.isnull().sum()
```

```python
df_1["Mileage"]=df_1["Mileage"].fillna(df_1["Mileage"].mean())
df_1["Engine"]=df_1["Engine"].fillna(df_1["Engine"].mode()[0])
df_1["Power"]=df_1["Power"].fillna(df_1["Power"].mean())
df_1["Seats"]=df_1["Seats"].fillna(df_1["Seats"].mode()[0])
```

```python
x1, y1 = df_1.drop("Price", axis=1), df_1["Price"]
y1
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
x2_imputed = pd.DataFrame(imputer.fit_transform(x2), columns=x2.columns)

model = LinearRegression()
model.fit(x1, y1)
y_pred = model.predict(x2_imputed)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y1, model.predict(x1))
mse = mean_squared_error(y1, model.predict(x1))
r2 = r2_score(y1, model.predict(x1))
mae, mse, r2
```
